##  Pattern matching pratice :

```python
text = ''' 
Hi, today is 17-Apr-2021, yesterday was 16-Apr-2021 and tomorrow will be 18-Apr-2021.
My schedule is free on 26-04-2021, 06.05.2021 and 16/Jun/2021.
You can reach out to me at myname2020@dummy.com or ask_help@demo.net & conference@demo.co.in
You can also call me at one of the following no's +6032-007 1212, +6099.100 3344, 017-9998800 etc.
'''
```

### Task1 : Match date pattern  dd-mmm-yyyy

```python
# create the pattern
pattern =  re.compile(r"\d{2}-[a-zA-Z]{3}-\d{4}") 


#find the pattern in text file 
print(pattern.findall(text))
```

### Task 2 : Find all the dates in the text 


```python
# create the pattern
pattern =  re.compile(r"\d{2}[-/.][a-zA-Z0-9]{2,3}[-/.]\d{4}") 


#find the pattern in text file 
print(pattern.findall(text))
```

### Task 3 : Find all the email IDs 

```python
email_pattern = re.compile(r'[a-zA-Z0-9]+@[a-zA-Z]+\.[a-z]+\S+')

email_pattern.findall(text)
```

### Task 4 : Find all phone numbers : 

```python
phone_pattern  = re.compile(r'(\+?)\d{3,}[-.]\d+(\s?)\d+')



[m.group() for m in phone_pattern.finditer(text)]
 
```

### Task 5 : text matching : Start with Hi and ends with !

```python
text2  ='Hi Good morning !!'

text_pattern = re.compile(r"^(H|h)i.*!$")

[tx.group() for tx in text_pattern.finditer(text2)]
```

### Task 6 : Subtitute particle letter or group of characters in a given string : 

```python
#hide salary amount with * or -

text3 = "My salary was $6000 in 2020 , now it is $120k in 2022"


salary_pattern =re.compile(r'\$\w+')

updated_text = salary_pattern.sub('****',text3)

print(updated_text)
```
